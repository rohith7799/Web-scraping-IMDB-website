```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/list/ls055386972/'
page = requests.get(url)
soup=BeautifulSoup(page.text,'html.parser') # creating soup object

big_list = soup.find('div', attrs = {'class': 'lister-list'})
big_list # big group
```

```python
s_no = []
Name = []
Year_of_Release = []
Rating = []

# serial number, name, year of release are all present in h3 tag

for x in big_list.find_all('h3'):
    s_no.append(x.find('span').text) # replace function to replace the .after the number
    Name.append(x.find('a').text)
    Year_of_Release.append(x.find('span', attrs = {'class' : "lister-item-year text-muted unbold"}).text)
    
# rating
for x in big_list.find_all('div', attrs = {'class':"ipl-rating-star small"}):
    Rating.append(x.find('span', attrs = {'class':"ipl-rating-star__rating"}).text)
```

```python
df = pd.DataFrame({'S_No': s_no,
                  'Name_Movie': Name,
                  'Release_year': Year_of_Release,
                  'Rating': Rating})
```

```python
df
```

```python
df['S_No'] = df['S_No'].apply(lambda x: x.replace('.', "")) # replacing . present in s_no column
df
```

```python
df['Release_year'] = df['Release_year'].apply(lambda x: x.replace('(', ""))
df['Release_year'] = df['Release_year'].apply(lambda x: x.replace(')', ""))
df['Release_year'] = df['Release_year'].apply(lambda x: x.replace('I', ""))
df
```

```python
# changing the datatypes of numeric columns
df['S_No'] = df['S_No'].astype('int')
df['Release_year'] = df['Release_year'].astype('int')
df['Rating'] = df['Rating'].astype('float')
```

```python
def top_50movies(url):
    import pandas as pd
    import requests
    from bs4 import BeautifulSoup
    url = url
    page = requests.get(url)
    soup=BeautifulSoup(page.text,'html.parser') # creating soup object
    big_list = soup.find('div', attrs = {'class': 'lister-list'})
    s_no = []
    Name = []
    Year_of_Release = []
    Rating = []

    # serial number, name, year of release are all present in h3 tag

    for x in big_list.find_all('h3'):
        s_no.append(x.find('span').text) # replace function to replace the .after the number
        Name.append(x.find('a').text)
        Year_of_Release.append(x.find('span', attrs = {'class' : "lister-item-year text-muted unbold"}).text)
    
    # rating
    for x in big_list.find_all('div', attrs = {'class':"ipl-rating-star small"}):
        Rating.append(x.find('span', attrs = {'class':"ipl-rating-star__rating"}).text)
        
    # creating dataframe
    df = pd.DataFrame({'S_No': s_no,
                  'Name_Movie': Name,
                  'Release_year': Year_of_Release,
                  'Rating': Rating})
    df['S_No'] = df['S_No'].apply(lambda x: x.replace('.', "")) # replacing . present in s_no column
    df['Release_year'] = df['Release_year'].apply(lambda x: x.replace('(', ""))
    df['Release_year'] = df['Release_year'].apply(lambda x: x.replace(')', ""))
    df['Release_year'] = df['Release_year'].apply(lambda x: x.replace('I', ""))
    # changing the datatypes of numeric columns
    df['S_No'] = df['S_No'].astype('int')
    df['Release_year'] = df['Release_year'].astype('int')
    df['Rating'] = df['Rating'].astype('float')
    
    # returning the data frame
    return df
```
